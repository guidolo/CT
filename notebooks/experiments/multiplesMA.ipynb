## Multiples MA


```python
import pandas as pd
from core.trade_service.instruments.moving_averages import MA
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('../../data/BTCUSDT-1h-data.csv')
```

```python
data.loc[:, 'timestamp'] = pd.to_datetime(data.timestamp)
```

```python
data = data.loc[(data.timestamp.dt.year==2020) & (data.timestamp.dt.month==7), ]
```

```python
ma_long = MA(data_column='close', period=25)
ma_short = MA(data_column='close', period=14)
```

```python
data.loc[:, 'ma_long'] = ma_long.evaluate(data)
data.loc[:, 'ma_short'] = ma_short.evaluate(data)
```

```python
data.loc[:,'close_untendence'] = data.close - data.ma_short
```

```python
data.close_untendence.plot()
```

```python
ma_long2 = MA(data_column='close_untendence', period=5)
ma_short2 = MA(data_column='close_untendence', period=3)
data.loc[:, 'ma_long2'] = ma_long2.evaluate(data)
data.loc[:, 'ma_short2'] = ma_short2.evaluate(data)
```

```python
start = 500
end = 1000
data.iloc[start:end].close.plot(color='black', figsize=(20,5))
data.iloc[start:end].ma_long.plot()
data.iloc[start:end].ma_short.plot()
plt.show()


data.iloc[start:end].close_untendence.plot(color='black', figsize=(20,5))
data.iloc[start:end].ma_long2.plot()
data.iloc[start:end].ma_short2.plot()
plt.legend()
```
